## Step 4: Data Cleaning

In this step:
- `Total Charges` is converted to numeric
- Missing values are handled
- Irrelevant columns such as `Churn Reason` are removed


```python
df_clean = df.copy()

drop_cols = [
    "CustomerID",
    "Count",
    "Country",
    "State",
    "City",
    "Zip Code",
    "Lat Long",
    "Latitude",
    "Longitude",
    "Churn Score",
    "CLTV"
]

df_clean.drop(columns=drop_cols, inplace=True)
df_clean.shape
```

```python
df_clean["Total Charges"] = pd.to_numeric(
    df_clean["Total Charges"], errors="coerce"
)

df_clean["Total Charges"].isna().sum()
```

```python
df_clean["Total Charges"].fillna(
    df_clean["Total Charges"].median(), inplace=True
)
```

Pandas raised a copy-on-write warning due to chained assignment with inplace operations.
I resolved it by explicitly reassigning the column, which is the recommended and safer approach.

```python
df_clean["Total Charges"] = df_clean["Total Charges"].fillna(
    df_clean["Total Charges"].median()
)
```

```python
df_clean["Total Charges"].isna().sum()
```

```python
df_clean.drop(columns=["Churn Reason"], inplace=True)
```

```python
df_clean.shape
```

```python
df_clean.drop(columns=["Churn Label"], inplace=True)
```

```python
df_clean.shape
```

Churn Value is the numeric representation of churn and is suitable as a target variable for classification models, whereas Churn Label is redundant.
